```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from qbstyles import mpl_style
import ing_theme_matplotlib.mpl_style as ing_style
```

```python
import seaborn as sns
```

```python
data = pd.read_csv('../data/Advertising.csv')
```

```python
data['total_spend'] = data['TV'] + data['radio'] + data['newspaper']
```

```python
mpl_style(True)
sns.regplot(data = data, x = 'total_spend', y = 'sales');
```

```python
ing_style(True)
sns.regplot(data = data, x = 'total_spend', y = 'sales');
```

```python
X = data['total_spend']
y = data['sales']
```

```python
lin_coefs = np.polyfit(X, y, deg = 1)
```

```python
potential_spend = np.linspace(0,500,130)
predicted_sales = lin_coefs[0] * potential_spend + lin_coefs[1]
```

```python
sns.scatterplot(data = data, x = 'total_spend', y = 'sales')
plt.plot(potential_spend, predicted_sales, color = 'red')
```

```python
lin_coefs_3 = np.polyfit(X, y, deg = 3)
predicted_sales = lin_coefs_3[0]*potential_spend**3 + lin_coefs_3[1]*potential_spend**2 + lin_coefs_3[2]*potential_spend + lin_coefs_3[3]
sns.scatterplot(data = data, x = 'total_spend', y = 'sales')
plt.plot(potential_spend, predicted_sales, color = 'red');
```
